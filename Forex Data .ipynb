### For Logistic Regression 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# X = feature values, all the columns except the last column
X = Time.iloc[:, :-1]

# y = target values, last column of the data frame
y = Time.iloc[:, -1]

# filter out the applicants that got admitted
TRUE = Time.loc[y == 1]

# filter out the applicants that din't get admission
FALSE = Time.loc[y == 0]

# plots
plt.scatter(TRUE.iloc[:, 0], TRUE.iloc[:, 1], s=5, label='TRUE')
plt.scatter(FALSE.iloc[:, 0], FALSE.iloc[:, 1], s=5, label='FALSE')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score 
model = LogisticRegression()
from sklearn.model_selection import train_test_split
model.fit(X, y)
```

```python
predicted_classes = model.predict(X)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(Time[['Timestamp', 'Bid_Open', 'Bid_High', 'Bid_Low', 'Bid_Close',
       'Bid_Volume', 'Ask_Open', 'Ask_High', 'Ask_Low', 'Ask_Close',
       'Ask_Volume']]
       ,Time['Class'], test_size=0.2,random_state=42,stratify=Time['Class'])
```

```python
vv = LogisticRegression().fit(X2_train,y2_train)
```

```python
yy = vv.predict(X2_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y2_train,yy)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
cm = confusion_matrix(y, predicted_classes)
print(cm)
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

Class=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(Class))
plt.xticks(Class, TRUE)
plt.yticks(Class, FALSE)
# create heatmap
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="gist_yarg" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.5)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```
